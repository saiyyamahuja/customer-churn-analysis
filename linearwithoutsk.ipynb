```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('lr.csv')
```

```python
x=np.array(df[['x']])
y=np.array(df[['y']])
plt.scatter(x,y)
plt.xlabel('Values of x')
plt.ylabel('Values of y')
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
x_std=np.std(x)
y_std=np.std(y)
```

```python
num=den=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print(f"Intercept {float(b0)}")
```

```python
y_pred=b0+b1*x
y_pred
```

```python
plt.plot(x,y_pred)
plt.scatter(x,y,c='red')
plt.show()
```

```python
#R-square=1-(SSR/SST)
num_r=0
den_r=0
for i in range(len(x)):
    num_r+=(y[i]-y_pred[i])**2
    den_r+=(y[i]-y_mean)**2
r_sq=1-(num_r/den_r)
print(f"The value of coefficient of determination R_square is {float(r_sq)}")
```
